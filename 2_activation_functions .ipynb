# Sigmoid the majorly used activation function for output layers in deep learning
# It gives value between 0 and 1

1_a04iKNbchayCAJ7-0QlesA.png

```python
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
sigmoid(-1)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
sigmoid(0.5)
```

# tanh
# tanh gives value between -1 and 1
Screen_Shot_2020-05-27_at_4.23.22_PM_dcuMBJl.png

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
tanh(1)
```

# Relu functions gives the value between 0 and 1

1_LiBZo_FcnKWqoU7M3GRKbA.png

```python
def relu(x):
    return max(0,x)
```

```python
relu(-100)
```

```python
relu(8)
```

# Leaky relu

leaky_relu_plot.png

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
